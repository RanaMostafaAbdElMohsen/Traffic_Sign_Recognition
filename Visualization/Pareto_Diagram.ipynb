```python
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerTuple
import matplotlib.patches as mpatches
```

```python
def ParetoDiagramGermanDataSet():
    colors = ['b', 'r', 'y', 'm', 'g']
    plt.title('Pareto Diagram for German Dataset')
    plt.xlabel('Model Parameters in Million (M)')
    plt.ylabel('Accuracy')

    STDNN = plt.scatter(14, 99.7, marker='x', color=colors[0])
    Ours = plt.scatter(3.2, 99.75, marker='x', color=colors[1])
    MCDNN  = plt.scatter(38.5, 99.5, marker='x', color=colors[2])
    MSCNN  = plt.scatter(1.43, 98.3, marker='x', color=colors[3])
    DHCNN = plt.scatter(6.26, 99.3, marker='x', color=colors[4]) #https://arxiv.org/pdf/1808.10524.pdf
#     OneCNN = plt.scatter(unknown, 98.17, marker='x', color=colors[4]) #https://www.fer.unizg.hr/_download/repository/ACPR_2015_JurisicFilkovicKalafatic.pdf
    
    plt.legend((STDNN, Ours, MCDNN, MSCNN, DHCNN),
               ('STDNN', 'Ours', 'MCDNN', 'MSCNN', 'DHCNN'),
               scatterpoints=1,
               loc='lower right',
               ncol=2,
               fontsize=10)

    plt.show()
```

```python
def ParetoDiagramBelgiumDataSet():
    colors = ['b', 'r', 'g']
    plt.title('Pareto Diagram for Belgium Dataset')
    plt.xlabel('Model Parameters in Million (M)')
    plt.ylabel('Accuracy')

    STDNN = plt.scatter(14, 98.87, marker='x', color=colors[0])
    Ours = plt.scatter(3.2, 99.25, marker='x', color=colors[1])
    DHCNN  = plt.scatter(6.26, 99.17, marker='x', color=colors[2])  #https://arxiv.org/pdf/1808.10524.pdf
    
    plt.legend((STDNN, Ours, DHCNN),
               ('STDNN', 'Ours','DHCNN'),
               scatterpoints=1,
               loc='upper right',
               ncol=2,
               fontsize=10)

    plt.show()
```

```python
def ParetoDiagramCombined():
    colors = ['b', 'r', 'y', 'm', 'g']
    marker=['x','o']
    plt.xlabel('Model Parameters in Million (M)')
    plt.ylabel('Accuracy')

    STDNN_German = plt.scatter(14, 99.7, marker='x', color=colors[0])
    STDNN_Belgium = plt.scatter(14, 98.87, marker='o', color=colors[0])
    Ours_German = plt.scatter(3.2, 99.75, marker='x', color=colors[1])
    Ours_Belgium = plt.scatter(3.2, 99.25, marker='o', color=colors[1])
    MCDNN_German  = plt.scatter(90, 99.5, marker='x', color=colors[2])
    MSCNN_German  = plt.scatter(1.43, 98.3, marker='x', color=colors[3])
    DHCNN_German = plt.scatter(6.26, 99.3, marker='x', color=colors[4]) #https://arxiv.org/pdf/1808.10524.pdf
    DHCNN_Belgium  = plt.scatter(6.26, 99.17, marker='o', color=colors[4]) #https://arxiv.org/pdf/1808.10524.pdf
    
    columns = [plt.plot([], [], marker[i], markerfacecolor='w',
                    markeredgecolor='k')[0] for i in range(2)]
    
    rows = [mpatches.Patch(color=colors[i]) for i in range(len(colors))]
    
    plt.legend(columns +rows,
               ["GTSRB", "BTSC"]+
               ["STDNN", "RIECNN","MCDNN","MSCNN", "DHCNN"]
               ,scatterpoints=1,
               loc='lower right',
               ncol=2,
               fontsize=10, numpoints=1,
               handler_map={tuple: HandlerTuple(ndivide=None)})
    
    plt.savefig('Combined_ParetoDiagram.png', bbox_inches='tight', dpi=300)
    plt.show()
```

```python
ParetoDiagramCombined()
```
