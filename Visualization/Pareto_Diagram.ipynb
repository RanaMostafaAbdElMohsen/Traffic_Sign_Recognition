```python
import matplotlib.pyplot as plt
```

```python
def ParetoDiagramGermanDataSet():
    colors = ['b', 'r', 'y', 'm']
    plt.title('Pareto Diagram for German Dataset')
    plt.xlabel('Model Parameters in Million (M)')
    plt.ylabel('Accuracy')

    STDNN = plt.scatter(14, 99.7, marker='x', color=colors[0])
    Ours = plt.scatter(3.2, 99.69, marker='x', color=colors[1])
    MCDNN  = plt.scatter(38.5, 99.5, marker='x', color=colors[2])
    MSCNN  = plt.scatter(1.43, 98.3, marker='x', color=colors[3])

    plt.legend((STDNN, Ours, MCDNN, MSCNN),
               ('STDNN', 'Ours', 'MCDNN', 'MSCNN'),
               scatterpoints=1,
               loc='lower right',
               ncol=2,
               fontsize=10)

    plt.show()
```

```python
def ParetoDiagramBelgiumDataSet():
    colors = ['b', 'r', 'y', 'm']
    plt.title('Pareto Diagram for Belgium Dataset')
    plt.xlabel('Model Parameters in Million (M)')
    plt.ylabel('Accuracy')

    STDNN = plt.scatter(14, 98.87, marker='x', color=colors[0])
    Ours = plt.scatter(3.2, 98.97, marker='x', color=colors[1])

    plt.legend((STDNN, Ours),
               ('STDNN', 'Ours'),
               scatterpoints=1,
               loc='lower right',
               ncol=2,
               fontsize=10)

    plt.show()
```

```python
ParetoDiagramGermanDataSet()
```

```python
ParetoDiagramBelgiumDataSet()
```
